```python
import pandas as pd
from random import randint
from bs4 import BeautifulSoup
from collections import Counter
```

```python
def cal():
    with open('test.html','r') as f:
        data = {}
        soup = BeautifulSoup(f,'html.parser')
        rows = soup.body.find_all("tr")
        for row in rows: 
            key = row.find_all('td')[0].text.strip()
            values = dict(Counter([ colour.strip() for colour in row.find_all('td')[1].text.split(",")]))
            data[key] = values
        return data
```

```python
frame = pd.DataFrame(cal().values(), index=['MONDAY','TUESDAY','WEDNESDAY','THURSDAY','FRIDAY']).fillna(0)
```
